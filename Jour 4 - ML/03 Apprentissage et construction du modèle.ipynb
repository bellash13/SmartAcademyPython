```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
#pip install scikit-learn
```

```python
chemin = 'assets/users-musics.csv'
data = pd.read_csv(chemin)
data.head()
```

```python
X = data.drop(columns=['Artiste'])
display(X)
```

DÃ©but d'apprentissage

```python
y = data['Artiste']

model = DecisionTreeClassifier()
model.fit(X, y)
 
```

```python
predictions = model.predict([[27,1]])
display(predictions)
```

```python
predictions = model.predict([[27,1],[27, 0],[20, 1],[37, 0],[24, 1],[22, 0]])
display(predictions)
```
