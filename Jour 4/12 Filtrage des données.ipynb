<h1>Filtrage des données</h1>

```python
import pandas as pd
chemin = 'assets/artistes_taxes.csv'
df = pd.read_csv(chemin)
```

```python
df.filter(items=['Mois','Taxe']) #revient à écrire df[['Mois','Taxe']]
```

```python
df.filter(like='Mois')
```

Vous pouvez également utiliser une expression régulière... Faire des recherches sur internet.
<code>df.filter(axis=0, regex='^5$')</code>
