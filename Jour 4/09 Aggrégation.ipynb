```python
age = df.Age
norm = (age -age.mean())/ age.std()
minMax = (age-age.min())/(age.max()-age.min())
minn =  age.min() 
maxx =  age.max() 
mean =  age.mean() 
std =  age.std() 

df['minMax'] = minMax
df['norm'] = norm
df['min'] = minn
df['max'] = maxx
df['mean'] = mean
df['std'] = std

df[['Artiste','Age','minMax', 'min', 'max','mean','std', 'norm']]
```
