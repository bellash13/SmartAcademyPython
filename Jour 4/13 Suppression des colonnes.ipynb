<h1>Suppression des colonnes</h1>

```python
import pandas as pd
chemin = 'assets/artistes_taxes.csv'
df = pd.read_csv(chemin)
df.head()
```

Supprimer la ligne dont la valeur est 0

```python
df.drop(0)
```

```python
df.head()
```

```python
df.drop(0, inplace=True)
df.head()
```

```python
df.head()
```

```python
df.drop(columns=['MoisChiffre'])
df.head()
```

```python
df.drop(columns=['MoisChiffre'], inplace=True)
df.head()
```

Pour supprimer une colonne, on met axis=1


```python
df.drop('Mois', axis=1, inplace=True)
df.head()
```
