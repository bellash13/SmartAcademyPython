```python
import pandas as pd
chemin = 'assets/artistes_taxes.csv'
df = pd.read_csv(chemin)
```

```python
df.groupby('Artiste').transform('nunique')
```

```python
df.groupby('Taxe').transform('nunique')
```

```python
df.groupby('Artiste')['Taxe'].transform('mean')
```
