```python
import pandas as pd
chemin = 'assets/artistes_taxes.csv'
df = pd.read_csv(chemin)
df.head()
```

```python
df.columns = [x.lower() for x in df.columns]
df.head()
```

```python
df.columns = [x.lower() for x in df.columns]
df.head()
```

Pour changer les noms des colonnes

```python
df.rename(columns={'mois': 'Month'})
#ou encore
df.rename({'taxe': 'Contribution'},axis=1)
df.head()
```

Vous remarquez qu'aucun changement n'est visible, 
c'est à cause de <code>inplace</code> qui est à <code>False</code> par défaut. 

```python
df.rename(columns={'mois': 'Month', 'moischiffre': 'MoisIndice'}, inplace=True)
#ou encore axis = 1 car axis = 0 par défaut
df.rename({'taxe': 'Contribution', 'artiste': 'STAR'},axis=1, inplace=True)
df.head()
```

On peut également utiliser l'expression lambda comme ci-dessous:

```python
df.rename(columns = lambda c: c.upper())
df.head()
```

```python
df.rename(columns = lambda c: c.upper(), inplace=True)
df.head()
```

```python
df.drop(0, inplace=True)
df.head()
```
